```python
import plotly.graph_objects as go
fig= go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'],y=혼인건수['2020'],name='2020'))
fig.add_trace(go.Bar(x=혼인건수['자치구별'],y=혼인건수['2022'],mode='iines',name='2022'))
fig.show()
```

```python
fig= go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'],y=혼인건수['2020'],name='2020'))
fig.add_trace(go.Scatter(x=혼인건수['자치구별'],y=혼인건수['2022'],mode='lines',name='2022'))
fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            buttons = list([
                dict(label="모두 그리기",method='update', args=[{'visible': [True, True]}]),
                dict(label="2020 그리기",method='update', args=[{'visible': [True, False]}]),
                dict(label="2022 그리기",method='update', args=[{'visible': [False, True]}])
            ])
        )

    ]
)
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(y=인구동향['출생아수 (명)'],x=인구동향['시점'], name='출생아수'))
fig.add_trace(go.Scatter(y=인구동향['혼인건수 (건)'],x=인구동향['시점'], mode='lines',name='혼인건수'))
fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            buttons = list([
               dict(label="모두 그리기", method='update', args=[{'visible': [True, True]}]),
                dict(label="출생아수 그리기", method='update', args=[{'visible': [True, False]}]),
                dict(label="혼인건수 그리기", method='update', args=[{'visible': [False, True]}])
            ])
        )
    ]
)
fig.show()
```
